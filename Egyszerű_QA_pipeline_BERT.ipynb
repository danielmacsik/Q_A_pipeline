# **Q&A Pipeline létrehozása**

```python
#itt megadhatjuk akár úgy a változót, hogy benne van az érték vagy bekérhetjük a felhasználótól is
#asked_question = "What are April's flowers?"
asked_question = input("What is your Question? ")
```

```python
from transformers import BertForQuestionAnswering
```

```python
model =BertForQuestionAnswering.from_pretrained('deepset/bert-base-cased-squad2')
#'deepset/bert-base-cased-squad2' - lesz a tesztben használva
# Azért BERT, mivel egyszerű és hatékony (teszteléshez, pedig megmutatja, az alapvető modellekkel mihez tudunk kezdeni)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('deepset/bert-base-cased-squad2')
#itt beállítom a tokenizerhez is a modellt
#egyébként lehetne a modellnevet is egy külön változóba kihelyezni, itt az egyszerűség miatt hagytam így
```

```python
from transformers import pipeline
nlp = pipeline('question-answering', model=model, tokenizer=tokenizer )
#pipeline kalibrálása
```

```python
#Válasz generálása a kérdésre
nlp({
    'question': asked_question,
    'context': long_text
})
```

```python
#Itt kiíratjuk a megoldást
#A válasz generálása a kérdésre
result= nlp({
    'question': asked_question,
    'context': long_text
})

print('The Question was:', asked_question)
print('The answer is:', result['answer'])
```
